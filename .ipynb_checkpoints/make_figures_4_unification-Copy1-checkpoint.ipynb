## Quad potential: Cooperativity, Conf Change, Fluct Change vs Force, Stiffness.

```julia
 = 1
N = 3
smi = 1
1 = 0.1

H = Matrix{Float64}(I, N,N)
h = [1, 0, 0]
hmag = 0
h1 = 0
f = zeros(N)
f[1] = 1
W = f*f'
fmag = LinRange(0, 1, 100)
Wmag = 10 .^ LinRange(-2,1, 100)

ddF = coopHelperQuad.([H], [h], [f], [f], [W], [W], , fmag, Wmag', hmag, 1, smi)'
dR  = confChangeHelperQuad.(1, h1, fmag, Wmag', , N)'
dFl = fluctChangeHelperQuad.(1, h1, fmag, Wmag', , N)'

fs = 17
fig, axs = subplots(2,2, figsize=(10,8))
fig.subplots_adjust(wspace = 0.5, hspace = 0.5)

ax = axs[1,2]
p = ax.pcolor(fmag, Wmag, ddF, cmap="inferno", linewidth=0, rasterized=true)
ax.set_xlabel("Ligand Force  "*L"||\mathbf{f}_a|| = ||\mathbf{f}_b||", fontsize=fs)
ax.set_ylabel("Ligand Stiffness   "*L"k_a = k_b", fontsize=fs)
ax.set_yscale("log")
clb = fig.colorbar(p, ax=ax)
clb.set_label("Cooperativity   "*L"\Delta \Delta F", fontsize=fs)

ax = axs[2,1]
p = ax.pcolor(fmag, Wmag, dR, cmap="inferno", linewidth=0, rasterized=true)
ax.set_xlabel("Ligand Force  "*L"||\mathbf{f}_a|| = ||\mathbf{f}_b||", fontsize=fs)
ax.set_ylabel("Ligand Stiffness   "*L"k_a = k_b", fontsize=fs)
ax.set_yscale("log")
clb = fig.colorbar(p, ax=ax)
clb.set_label("Conformational change \n"*L"|| \langle \mathbf{r} \rangle_{11} - \langle \mathbf{r} \rangle_{00} ||", fontsize=fs)

ax = axs[2,2]
p = ax.pcolor(fmag, Wmag, abs.(dFl), cmap="inferno", linewidth=0, rasterized=true)
ax.set_xlabel("Ligand Force  "*L"||\mathbf{f}_a|| = ||\mathbf{f}_b||", fontsize=fs)
ax.set_ylabel("Ligand Stiffness   "*L"k_a = k_b", fontsize=fs)
ax.set_yscale("log")
clb = fig.colorbar(p, ax=ax)
clb.set_label("Fluctuation change \n"*L"\left|\sqrt{\langle \mathbf{r}^2 \rangle_{11} - \langle \mathbf{r} \rangle_{11}^2} - \sqrt{\langle \mathbf{r}^2 \rangle_{00} - \langle \mathbf{r} \rangle_{00}^2}\right|", fontsize=fs)

axs[1].tick_params(bottom=false, top=false, left=false, right=false) # labelbottom=false, labelleft=false)
axs[1].tick_params(bottom=false, top=false, left=false, right=false) # labelbottom=false, labelleft=false)
axs[1].spines["top"].set_visible(false)
axs[1].spines["right"].set_visible(false)
axs[1].spines["bottom"].set_visible(false)
axs[1].spines["left"].set_visible(false)
axs[1].set_xticklabels([])
axs[1].set_yticklabels([]);
#fig.savefig("LinearModels_forceStiffness.svg", bbox_inches="tight")
```

## Fig 3: Bistable potential: Coop as a function of $\lambda_1$ and $\Vert g\Vert$ for different conditions.

```julia
## Soft Mode, Added Spring.
 = 1
N = 3
smi = 3

H = Matrix{Float64}(I, N,N)
H .*= 5
h = normalize(-[1, 0, 0.1])
g = normalize([1, 0, 0])
fa = zeros(N)
fa[1] = 1
fa[2] = 0
fa[smi] = 0.1
normalize!(fa)
fb= zeros(N)
fb[1] = 1
fb[2] = 0
fb[smi] = 0.1
normalize!(fb)
Wa = diagm(fa .^ 2)
Wb = diagm(fb .^ 2)
1 = 10 .^ LinRange(-2,0,100)
gmag = LinRange(0,10,100)

function plotter(ax, x, y, z,fs; title="", clbLabel="", vmax=maximum(z))
    p = ax.pcolor(x, y, z, cmap="inferno", linewidth=0, rasterized=true, vmax=vmax)
#     p = ax.contourf(x, y, z, 30, cmap="inferno", vmax=vmax)
#     pa = ax.contour(x, y, z, 30, linewidths=1, vmax=vmax)
    ax.set_ylabel("Soft Mode Stiffness "*L"\lambda_1", fontsize=fs)
    ax.set_xlabel("Conf. Switch Size "*L"\Vert \mathbf{g} \Vert", fontsize=fs)
    ax.set_yscale("log")
    ax.set_title(title, fontsize=fs)
    clb = fig.colorbar(p, ax=ax)
    clb.set_label(clbLabel, fontsize=fs)
end

fs = 20
fig, axs = subplots(3,2, figsize=(10, 13))
fig.subplots_adjust(wspace=0.4, hspace=0.3)

hmag = 1
fmag = 1
Wmag = 0
ddF1 = coopHelperBi.([H], [h], [g], [fa], [fb], [Wa], [Wb],
    , fmag, Wmag, hmag,gmag, 1', smi)'
CC1 = confChangeHelperBi.([H], [h], [g], [fa], [fb], [Wa], [Wb],
    , fmag, Wmag, hmag,gmag, 1', smi)'
FC1 = fluctChangeHelperBi.([H], [h], [g], [fa], [fb], [Wa], [Wb],
    , fmag, Wmag, hmag,gmag, 1', smi)'
ax = axs[1,1]
plotter(ax, gmag, 1, ddF1, fs, clbLabel="Cooperativity "*L"\Delta \Delta F")
ax = axs[2,1]
plotter(ax, gmag, 1, CC1, fs, clbLabel="Conformational Change "*L"\phi", vmax=8)
ax = axs[3,1]
plotter(ax, gmag, 1, FC1, fs, clbLabel="Fluctuation Change "*L"\psi", vmax=1)

#############################
hmag = 0
fmag = 0
Wmag = 1

ddF2 = coopHelperBi.([H], [h], [g], [fa], [fb], [Wa], [Wb],
    , fmag, Wmag, hmag, gmag, 1', smi)'
CC2 = confChangeHelperBi.([H], [h], [g], [fa], [fb], [Wa], [Wb],
    , fmag, Wmag, hmag, gmag, 1', smi)'
FC2 = fluctChangeHelperBi.([H], [h], [g], [fa], [fb], [Wa], [Wb],
    , fmag, Wmag, hmag, gmag, 1', smi)'
ax = axs[1,2]
plotter(ax, gmag, 1, ddF2, fs, clbLabel="Cooperativity "*L"\Delta \Delta F")
ax = axs[2,2]
plotter(ax, gmag, 1, CC2, fs, clbLabel="Conformational Change "*L"\phi",vmax=1)
ax = axs[3,2]
plotter(ax, gmag, 1, FC2, fs, clbLabel="Fluctuation Change "*L"\psi")

colorbar_ax = fig.axes[8]
newlabels = string.(1:8)
newlabels[end] = "\$>8\$"
colorbar_ax.set_yticklabels(newlabels);
fig.savefig("./Bistable_model/components/BistableModel_force_and_stiffness.svg", bbox_inches="tight")
```

```julia
# work from before the Gordon Conference.

function makeAB(H)
    A = -copy(H)
    A[:,1] .= 0
    A[1,1] = 1
    B = -Matrix{Float64}(I(size(H,1)))
    B[:,1] .= H[:,1]
    return A, B
end

function Fswitch(H, , gmag::Number, h, fa, fb)
    @assert size(H,1) == length(fa) == length(fb) == length(h)
    
    N = length(h)
    e1 = zeros(N)
    e1[1] = 1
    g = gmag * e1
    Hbar = H[2:end, 2:end] 
    A, B = makeAB(H)
    h = copy(h)
    h .+=  fa .+ fb
    
    # left half integral
    t1 = 0.5 * sqrt( (2 * pi / )^(N-1)/ det(Hbar) )
    t2 = sqrt(2 * pi / ( * e1' * inv(A) * B * e1 ) )
    t3 = sqrt(   * e1' * inv(A) * B * e1 / 2 )
    t4 = H[1,1] * g[1] -  H[1,:]' * h
    t5 = exp(  * (h-g)' * inv(H) * (h-g) / 2)
    Ia = t1 * t2 * (erf(t3 * t4) + 1) * t5
    Ja = 2 *t1 * t2 * t5
    
    # right half integral
    t4 = - H[1,1] * g[1] -  H[1,:]' * h
    t5 = exp(  * (h+g)' * inv(H) * (h+g) / 2)
    Ib = t1 * t2 * (1 - erf(t3 * t4)) * t5
    
    return - log(Ia + Ib) / 
end

function coopSwitch(H::Matrix, , gmag::Number, h, fa, fb)
    N = length(h)
    zeroVec = zeros(N)
    F00 = Fswitch(H, , gmag, h, zeroVec, zeroVec)
    F10 = Fswitch(H, , gmag, h, fa, zeroVec)
    F01 = Fswitch(H, , gmag, h, zeroVec, fb)
    F11 = Fswitch(H, , gmag, h, fa, fb)
    return F10 + F01 - F00 - F11
end

function coopSwitch(H11::Number, , gmag::Number, h, fa, fb)
    N = length(h)
    H = Matrix{Float64}(I, N, N)
    H[3,3] = H11
    if norm(g) == 0
        h = float.([0, 0, 0])
    else
        h = -sqrt(2)*norm(fa) * g / (2*norm(g))
    end
    return coopSwitch(H, , gmag::Number, h, fa, fb)
end

function Fu(H::Vector, Wa::Vector, Wb::Vector, , , )
    N = length(Wa)
    H0 =  * ones(N)
    Ffold = - log((2*pi/)^N / prod( H + Wa + Wb)) /  + 
    Wa = copy(Wa)
    Wa[1] = 0
    Wb = copy(Wb)
    Wb[1] = 0
    Funfold = - log((2*pi/)^N / prod( H0 + Wa + Wb)) / 
    return - log( exp(-* Ffold) + exp(- * Funfold) ) / 
end

function coopFold(H::Vector, Wa::Vector, Wb::Vector, , , )
    N = length(Wa)
    zeroVec = zeros(N)
    F00 = Fu(H, zeroVec, zeroVec, , , )
    F10 = Fu(H,  Wa, zeroVec, , , )
    F01 = Fu(H, zeroVec, Wb, , , )
    F11 = Fu(H, Wa, Wb, , , )
    return F10 + F01 - F00 - F11
end

function coopFold(H11::Number, Wa::Vector, Wb::Vector, , , )
    N = length(Wa)
    H = ones(N)
    H[1,1] = H11
    return coopFold(H, Wa, Wb, , , )
end

function Eswitch(r, H, g, h, fa, fb)
    return r' * H * r - abs( g' * r) - (h + fa + fb)' * r
end

function Fef(r, H, Wa, Wb, , , )
    
    HH = zeros(length(H), length(H))
    HH[diagind(HH)] = H
    
    HH0 =  zeros(length(H), length(H))
    HH0[diagind(HH0)] =  * ones(N)
    
    WWa =  zeros(length(H), length(H))
    WWa[diagind(WWa)] = Wa
    
    WWb =  zeros(length(H), length(H))
    WWb[diagind(WWb)] = Wb
    
    dett1 = prod(diag(HH + WWa + WWb)[2:end])
    Ef = r' * HH * r + r' * (WWa + WWb) * r + 
    
    WWa = copy(WWa)
    WWa[1] = 0
    WWb = copy(WWb)
    WWb[1] = 0
    
    
    Eu = r' * HH0 * r + r' * (WWa + WWb) * r
    dett2 = prod(diag(HH0 + WWa + WWb))
    return - log( sqrt(1/dett1)*exp(-* Ef) + sqrt(1/dett2)*exp(-*Eu) ) / 
end
```
