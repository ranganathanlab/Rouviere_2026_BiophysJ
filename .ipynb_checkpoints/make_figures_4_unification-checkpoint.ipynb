## Fig 3: Bistable potential: Coop as a function of $\lambda_1$ and $\Vert g\Vert$

```julia
 = 1
N = 3
smi = 2
 = 0.1
O1 = 1
O2 = 0.1

# parameters in the enegy landscape using mode cooridnates.
H = 10*diagm(ones(N)); H[smi, smi] = 
g = [1,0,0]
h = [-1,0,0] * 0.7035975

fa_pre = [1,0,0]
fb_pre = [0,1,0]
_, Q, _ =  prepareSystem(O1, O2, )
fa = Q'fa_pre
fb = Q'fb_pre
Wa = fa*fb'
Wb = fb*fb'

hmag, fmag, Wmag = 1, 1, 0 
 = 10 .^ LinRange(-2,0,100)
gmag = LinRange(0,10,100)


function plotter(ax, x, y, z,fs; title="", clbLabel="", vmax=maximum(z))
    p = ax.pcolor(x, y, z, cmap="inferno", linewidth=0, rasterized=true, vmax=vmax)
    ax.set_ylabel("Soft Mode Stiffness "*L"\lambda_1", fontsize=fs)
    ax.set_xlabel("Conf. Switch Size "*L"\Vert \mathbf{g} \Vert", fontsize=fs)
    ax.set_yscale("log")
    ax.set_title(title, fontsize=fs)
    clb = fig.colorbar(p, ax=ax)
    clb.set_label(clbLabel, fontsize=fs)
end

fs = 20
fig, axs = subplots(2,3, figsize=(15, 8))
fig.subplots_adjust(wspace=0.4, hspace=0.3)

ddF1 = coopHelperBi.([H], [h], [g], [fa], [fb], [Wa], [Wb],
    , fmag, Wmag, hmag,gmag, ', smi)'
CC1 = confChangeHelperBi.([H], [h], [g], [fa], [fb], [Wa], [Wb],
    , fmag, Wmag, hmag,gmag, ', smi)'
FC1 = fluctChangeHelperBi.([H], [h], [g], [fa], [fb], [Wa], [Wb],
    , fmag, Wmag, hmag,gmag, ', smi)'

ax = axs[1,1]
plotter(ax, gmag, , ddF1, fs, clbLabel="Cooperativity "*L"\Delta \Delta F")
ax = axs[1,2]
plotter(ax, gmag, , CC1, fs, clbLabel="Conformational Change "*L"\phi", vmax=5)
ax = axs[1,3]
plotter(ax, gmag, , FC1, fs, clbLabel="Fluctuation Change "*L"\psi", vmax=1)

gmag = 10
out = coopEachConfHelper.([H], [h], [g], [fa], [fb], , fmag, hmag, gmag, , smi)
ddF_a = getindex.(out, 1)
ddF_b = getindex.(out, 2);
ddF = coopHelperBi.([H], [h], [g], [fa], [fb], [Wa], [Wb], , fmag, Wmag, hmag,gmag, , smi)

ax = axs[2,3]
ax.plot(, ddF_a, c="g", lw=4)
ax.plot(, ddF_b, c="r", lw=2)
ax.plot(,   ddF, c="b", lw=2)
ax.set_xscale("log")
ax.set_xlabel("Soft Mode Stiffness "*L"\lambda_1", fontsize=fs)
ax.set_xlabel("Soft Mode Stiffness "*L"\lambda_1", fontsize=fs)
ax.set_ylabel("Cooperativity", fontsize=fs)
ax.legend(["State 1", "State 2", "Both States"])

axs[2,1].tick_params(bottom=false, top=false, left=false, right=false) # labelbottom=false, labelleft=false)
axs[2,1].tick_params(bottom=false, top=false, left=false, right=false) # labelbottom=false, labelleft=false)
axs[2,1].spines["top"].set_visible(false)
axs[2,1].spines["right"].set_visible(false)
axs[2,1].spines["bottom"].set_visible(false)
axs[2,1].spines["left"].set_visible(false)
axs[2,1].set_xticklabels([])
axs[2,1].set_yticklabels([])

axs[2,2].tick_params(bottom=false, top=false, left=false, right=false) # labelbottom=false, labelleft=false)
axs[2,2].tick_params(bottom=false, top=false, left=false, right=false) # labelbottom=false, labelleft=false)
axs[2,2].spines["top"].set_visible(false)
axs[2,2].spines["right"].set_visible(false)
axs[2,2].spines["bottom"].set_visible(false)
axs[2,2].spines["left"].set_visible(false)
axs[2,2].set_xticklabels([])
axs[2,2].set_yticklabels([])

fig.savefig("./Bistable_model/components/BistableModel_force_and_stiffness.svg", bbox_inches="tight")
```
