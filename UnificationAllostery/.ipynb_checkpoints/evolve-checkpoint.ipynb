```julia
## Testing Different ways to build a hessian.
W = 7
L = 7
offset = 0.2
β = 1
ζ = 1e7

# get Structure matrix and Hessian
r, A = buildStructure(W, L, offset);
bondList = findall(x -> x != 0, LowerTriangular(A) )
numBonds = length(bondList)
K = genRandK(numBonds)
S = computeStructureMatrix(r, A)
h = zeros(size(S,2))
H = S' * diagm(K) * S
G = pinv(H)

bonda = 13
bondb = numBonds - 4
ba, bb = S[bonda,:], S[bondb,:]
k0 = 1000
l0 = 0
dk = 0
dl = 1
numIter = 10000
net0 = Network(K, pinv(H), similar(H), similar(H), zeros(size(H,1)), [logpdet(H)]);

ddF = computeCoopQuadFast_kl(net0, h, ba, bb, k0, l0, dk, dl, β)
```

```julia
k0= 1e7
B = k0*(ba*ba' + bb * bb')
P = pinv(H + B)

Aa = G*ba*ba'*G
Ab = G*bb*bb'*G
Oa = ba'*G*ba
Ob = bb'*G*bb
Oab = ba'*G*bb


T1 = G - (1 + Oab^2/(Oa*Ob)) * (Aa/Oa + Ab/Ob)
T2 = T1 + (G*ba*ba'*G*bb*bb'*G + G*bb*bb'*G*ba*ba'*G) / (Oa*Ob)


(G*ba*ba'*G)/(1+ba'*G*ba) - (G*bb*bb'*G)/(1 + bb'*G*bb)
@show norm(P - T1, Inf)
@show norm(P - T2, Inf)



# U = P - O
# F = G*ba*ba'*G*bb*bb'*G + (G*ba*ba'*G*bb*bb'*G)'
# V = U - 0.2781536688872516 * F
```
