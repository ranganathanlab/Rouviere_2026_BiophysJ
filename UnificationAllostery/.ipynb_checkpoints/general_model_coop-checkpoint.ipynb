```julia
# Energy vs conf changes

N = 3
 = 1
H = Matrix{Float64}(I(N))
H11 = 1
gmag = 1
g = float.([gmag, 0, 0])
fa = float.([1,  1, 0.1])
fb = float.([1, -1, 0.1])

if norm(g) == 0
    h = float.([-1, 0, 0])
else
    h = -sqrt(2)*norm(fa) * g / (2*norm(g))
end

r00 = inv(H) * (h - g)
r11 = inv(H) * (h + fa + fb + g)
dr = r11 - r00

n = 100
e00 = zeros(n)
e10 = zeros(n)
e01 = zeros(n)
e11 = zeros(n)
for i in 1:n
    r = r00 + (i/n) * dr
    e00[i] = Eswitch(r, H, g, h, zeros(3), zeros(3))
    e10[i] = Eswitch(r, H, g, h, fa, zeros(3))
    e01[i] = Eswitch(r, H, g, h, zeros(3), fb)
    e11[i] = Eswitch(r, H, g, h, fa, fb)
end


fs = 14
fig, ax = subplots(1,1, figsize=(3, 2.2))
fig.subplots_adjust(top=0.9, wspace=0.4, hspace=0.4)
x = LinRange(0,1, n)
ax.plot(x, e00, label="00")
ax.plot(x, e10, label="10")
ax.plot(x, e01, label="01")
ax.plot(x, e11, label="11")
#ax.set_title(L"k_m=0, \ell_m=0, \epsilon=-50")
ax.set_xlabel("Conf. Coordinate  "*L"y", fontsize=fs)
ax.set_ylabel("Energy  "*L"U(y)", fontsize=fs)
ax.legend()
#fig.savefig("1d_energies_conf_switch.svg",bbox_inches="tight")
```

## Plotting Contours of bistable landscape.

```julia
function V(x,y, H::Matrix, g::Vector,  h::Vector)
    r = [x, y]
    return  r' * H * r - abs(g' * r) - h' * r
end
```

```julia
x = LinRange(-3, 3, 200)
y =  LinRange(-3, 3, 100)

H = [1 +0.5; +0.5 1]
g = [1,0]
h = [0,1]

z = V.(x, y', [H], [g], [h])';

contour(x,y, z, 100)
#axis("equal")
axis("square")
xlabel(L"x", fontsize=20)
ylabel(L"y", fontsize=20)
colorbar()
```
