```julia
function coop_IF_num(H, fa, fb)
    E10 = - 0.5 * fa' * pinv(H) * fa
    E01 = - 0.5 * fb' * pinv(H) * fb
    E11 = - 0.5 * (fa + fb)' * pinv(H) * (fa + fb)
    return E10 + E01 - E11
end
function coop_IF_theory(H, fa, fb)
    return fa' * pinv(H) * fb
end

function coop_D_num(H, Wa, Wb,  β)
    function F(H, β)
        det = prod(eigvals(H)[4:end])
        return -log( (2 * pi / β) / det ) / (2β)
    end
    F00 = F(H, β)
    F10 = F(H + Wa, β)
    F01 = F(H + Wb, β)
    F11 = F(H + Wa + Wb, β)
    return F10 + F01 - F00 - F11
end
function coop_D_theory(H, S, bondsite_a, bondsite_b, dka, dkb, β)
    ea = zeros(size(S,1))
    ea[bondsite_a] = 1
    eb = zeros(size(S,1))
    eb[bondsite_b] = 1
    values = eigvals(H)
    Q = eigvecs(H)
    Λ = diagm(0 => eigvals(H))
    J = inv(Λ)
    J[1,1] = J[2,2] = J[3,3] = 0
    ca = Q' * S' * ea
    cb = Q' * S' * eb
    num1 = dka * dkb * (ca' * J * cb)^2
    d1 = 1 + dka * ca' * J * ca
    d2 = 1 + dkb * cb' * J * cb
    return - log(1 - num1/(d1*d2))/(2*β)
end

function coop_B_num(H, fa, fb, Wa, Wb, β)    
    function F(H::Matrix, h::Vector, β::Number)
        N = size(H,1)
        E = - 0.5 * h' * pinv(H) * h # ground state energy 
        det = prod(eigvals(H)[4:end])
        Fhar = -log( (2 * pi / β)^N / det ) / (2β)
        return E + Fhar
    end
    zeroVec = zeros(size(fa))
    F00 = F(H, zeroVec,  β)
    F10 = F(H + Wa, fa, β)
    F01 = F(H + Wb, fb, β)
    F11 = F(H + Wa + Wb, fa+fb, β)
    return F10 + F01 - F00 - F11
end
function coop_B_num(H, S, fa, fb, bondsite_a, bondsite_b, dka, dkb, β)
    numBonds = size(S,1)
    Ka = zeros(numBonds, numBonds)
    Ka[bondsite_a, bondsite_a] = dka
    Kb = zeros(numBonds, numBonds)
    Kb[bondsite_b, bondsite_b] = dkb
    Wa = S' * Ka * S
    Wb = S' * Kb * S;
    return coop_B_num(H, fa, fb, Wa, Wb, β)   
end

function coop_B_theory_1term(H, S, fa, fb, bondsite_a, bondsite_b, dka, dkb, β)
    # return the first order correction term to the 
    # cooperativity.
    ea = zeros(size(S,1))
    ea[bondsite_a] = 1
    eb = zeros(size(S,1))
    eb[bondsite_b] = 1
    values = eigvals(H)
    Q = eigvecs(H)
    Λ = diagm(0 => eigvals(H))
    J = inv(Λ)
    J[1,1] = J[2,2] = J[3,3] = 0
    ca = Q' * S' * ea
    cb = Q' * S' * eb    
    va = Q' * fa
    vb = Q' * fb
    t1 = -0.5 * (va + vb)' * ( J * (dka*ca*ca' + dkb*cb*cb') * J) * (va + vb)
    t2 = 0.5 * dka * (va' * J * ca )^2
    t3 = 0.5 * dkb * (vb' * J * cb )^2
    return t1 + t2 + t3
end
function coop_B_theory(H, S, fa, fb, bondsite_a, bondsite_b, dka, dkb, β)
    ddF_IF = coop_IF_theory(H, fa, fb)
    ddF_D = coop_D_theory(H, S, bondsite_a, bondsite_b, dka, dkb, β)
    C = coop_B_theory_1term(H, S, fa, fb, bondsite_a, bondsite_b, dka, dkb, β)
    return ddF_IF + ddF_D + C
end
function coop_B_simple(H, S, fa, fb, bondsite_a, bondsite_b, dka, dkb, β)
    ddF_IF = coop_IF_theory(H, fa, fb)
    ddF_D = coop_D_theory(H, S, bondsite_a, bondsite_b, dka, dkb, β)
    return ddF_IF + ddF_D
end
```

## Validate the separated  quadratic form.

$\mathbf{x}^\top C  \mathbf{x} = (C_{11} - \mathbf{t}^\top \bar{C}^{-1} \mathbf{t}) x_1^2 + (\bar{\mathbf{x}} + x_1 \bar{C}^{-1} \mathbf{t} )^\top \bar{C} (\bar{\mathbf{x}} + x_1 \bar{C}^{-1} \mathbf{t} )$

We have assumed $C$ is symmetric and invertable. look in note 20 for definitions

```julia
qform(x, C) = x' * C * x

function sqform(x, C)
    C11 = C[1,1]
    x1 = x[1]
    t = C[2:end, 1]
    Cbar = C[2:end, 2:end]
    xbar = x[2:end]
    B = pinv(Cbar)
    z = xbar + x1 * B * t
    
    t1 = (C11 - t' * B * t) *x1^2
    t2 = z' * Cbar * z
    return t1 + t2
end

function sqform2(x, C)
    C11 = C[1,1]
    x1 = x[1]
    t = C[2:end, 1]
    Cbar = C[2:end, 2:end]
    xbar = x[2:end]
    B = pinv(Cbar)
    
    t1 = C11*x1^2 
    t2 = 2* x1 * t' * xbar
    t3 = xbar' * Cbar * xbar
    return t1 + t2 + t3
end

N = 10
A = randn(N, N)
C = A + A'
x = rand(N)

@show qform(x, C)
@show sqform(x, C)
@show sqform2(x, C)
```

## Validate the stiff spring limit of cooperativity under pure stiffening. 

```julia
function dynamicFrustration(G, ba, bb, h)
    a = ba'*G*ba
    b = bb'*G*bb
    c = ba'*G*bb
    ha = h'*G*ba
    hb = h'*G*bb
    
    t1 = c^2 / ( a*b - c^2)
    t2 = ha^2 / a
    t3 = hb^2/b
    t4 = 2 *ha*hb/c
    return -t1*(t2+t3-t4) / 2
end
```
