```julia
using PyPlot, LinearAlgebra, LaTeXStrings, SpecialFunctions, Revise, StatsBase
push!(LOAD_PATH,"./src")
using Cooperativity, LinearElastic
```

```julia
function run_simulation()
    
    N = 3
    r0 = zeros(N)
    H = diagm(ones(N))
    H = [1 -0.5 0; 
         -0.5 1 0;
         0 0 1]
    H0 = 0.01 * diagm(ones(N))
    g = [0,0,0]
    h = [0,0,0]
    ϵ = 7
    β = 1
    computeEnergy(r) = U_full(r, H, H0, g, h, ϵ, β)

    r0 = zeros(N)
    n_iter = 2_000_000
    γ = 0.5
    β_mcmc = 1
    t_sample = 200
    return run_mcmc(computeEnergy, r0, n_iter, γ, β_mcmc, t_sample)
end

# collective variables.#############################
Zcc(r::Vector, ΔR::Vector) = r'ΔR
function Zrmsd(r::Vector, R_nat::Vector)
    a = 1 / sqrt(length(r))
    return a * norm(r - R_nat)
end
```

```julia
@time energies, confs = run_simulation();

### Analysis #############################

fig,ax=subplots(figsize=(13,3))
ax.plot(energies, lw=0.5)

fig, ax = subplots(figsize=(4,4))
x, y, z = confs[1,:], confs[2,:], confs[3,:]
ax.scatter(x,y, s=1)

ΔR = [1,0,0]
q = mapslices(x -> Zcc(x, ΔR), confs, dims=1)';
fig, ax = subplots(figsize=(4,3))
ax.hist(q, 50)

R_nat = [0,0,0]
Q = mapslices(x -> Zrmsd(x, R_nat), confs, dims=1)';
fig, ax = subplots(figsize=(4,3))
ax.hist(sqrt.(Q), 50);
```

```julia
function make_2d_FreeEnergy(x, y, β, x_edges, y_edges)
    data = (x,y)
    edges = (x_edges, y_edges)
    h = fit(Histogram, data, edges)
    P = Matrix{Float64}(h.weights)
    P ./= sum(P)
    for i in eachindex(P)
        P[i] == 0 && (P[i] = NaN)
    end
    return - log.(P) ./ β # free energy
end


x = randn(100000)
y = randn(100000)
x_edges = LinRange(-4, 4, 20+1)
y_edges = LinRange(-4, 4, 20+1)

F = make_2d_FreeEnergy(x, y, β, x_edges, y_edges)

contourf(x_edges[1:end-1],y_edges[1:end-1], F)
```

## Minimal Analytical Unfolding Model Landscape.

```julia
F1(x,N, k) = -log(x^(3N-1)) + k*x^2
F2(x, N, k, k0, e) = -log( x^(3N-1)*exp(-(k+k0)*x^2 - e) + x^(3N-1)*exp(-k0*x^2) )
#F2(x, N, k, k0, e) = -log( x^(3N-1)*exp(-(k+k0)*x^2 - e) )
```

```julia
x = LinRange(0, 30, 1000);
k, k0 = 1, 0.05
N = 5
e = - 30


fig, ax = subplots(figsize=(4,3))
ax.plot(x, F1.(x, N, k) .+ e, lw=2, c="r" )
ax.plot(x, F2.(x, N, k, k0, e), lw=2, c="b")
ax.set_ylim([-50,50])
ax.set_xlabel(L"Q", fontsize=14)
ax.set_ylabel(L"\tilde{F}_{\rm RMSD}(Q)", fontsize=14)
ax.legend(["Quad Landscape","Coop Unfolding"])
fig.savefig("Free_energy_unfolding.pdf", bbox_inches="tight")
```
